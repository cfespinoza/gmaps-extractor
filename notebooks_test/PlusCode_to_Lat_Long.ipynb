```python
import requests
import json
base_url = "https://plus.codes/api?address={localidad}&ekey={api_key}&email=misco"
# api_key = "misco"
l_info = {}
for l in localidades:
    l_enriquecido = {}
    c_localidad = l.get("localidad")
    url = base_url.format(localidad=c_localidad, api_key=api_key)
    response = requests.get(url)
    parsed_responsed = json.loads(response.text)
    # todo control errors
    l_enriquecido["full_info"] = parsed_responsed
    l_enriquecido["cuadrante"] = parsed_responsed.get("plus_code").get("global_code")
    l_enriquecido["cuadrante_base"] = l_enriquecido["cuadrante"][:4]
    if l_info.get(c_localidad):
        print("Localidad {loc} ya registrada".format(loc=c_localidad))
    else:
        l_info[c_localidad] = l_enriquecido

l_info
```
